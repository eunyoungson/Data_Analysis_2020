```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
vix =iris[iris.species == 'virginica'].sepal_length
viy =iris[iris.species == 'virginica'].sepal_width
vix1 =iris[iris.species == 'virginica'].petal_length
viy1 =iris[iris.species == 'virginica'].petal_width
plt.boxplot([vix,viy,vix1,viy1],labels=['sepal_length','sepal_width','petal_length','petal_width'])

plt.show()
```

```python

fig, axes = plt.subplots(1, 3,figsize = (15,8), sharey=True)
for i, spec in enumerate (['setosa', 'versicolor', 'virginica']):
    sl = iris[iris.species == spec].sepal_length
    sw = iris[iris.species == spec].sepal_width
    pl = iris[iris.species == spec].petal_length
    pw = iris[iris.species == spec].petal_width
    ax = axes[i]
    ax.boxplot((sl, sw, pl ,pw), labels=['sepal_length','sepal_width','petal_length','petal_width'])
    ax.set_title(f"{spec} 품종")
    fig.suptitle("species")
plt.show()
```

```python
ans  = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

```python
ds1 = ans[ans['dataset'] == 'I'];    ds2 = ans[ans['dataset'] == 'II'] 
ds3 = ans[ans['dataset'] == 'III'];  ds4 = ans[ans['dataset'] == 'IV']
```

```python
fig = plt.figure(figsize=(10,6)) 
ax1 = fig.add_subplot(2, 2, 1);  ax2 = fig.add_subplot(2, 2, 2) 
ax3 = fig.add_subplot(2, 2, 3);    ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(ds1['x'], ds1['y'], 'o'); ax2.plot(ds2['x'], ds2['y'], 'o') 
ax3.plot(ds3['x'], ds3['y'], 'o'); ax4.plot(ds4['x'], ds4['y'], 'o') 

ax1.set_title('Dataset 1');  ax2.set_title('Dataset 2')
ax3.set_title('Dataset 3');  ax4.set_title('Dataset 4')

fig.suptitle("Anscombe's Quartet")
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2,figsize = (15,8), sharey = True,sharex = True)

for i,num in enumerate(['I','II','III','IV']) :   
    #ax =axes[i] 
    #ds= ds[i]
    ds = ans[ans.dataset == num] 
    x =ds.x ; y =ds.y
    ax = axes[i//2][i%2]
    ax.plot(x, y, 'o',markersize =10)
    ax.plot([3,20],[4.5,13],'r-',linewidth=2)
    ax.set_title(f'Dataset {num}')

fig.suptitle("Anscombe's Quartet",fontsize =20, y =1.05)
fig.tight_layout()
```
