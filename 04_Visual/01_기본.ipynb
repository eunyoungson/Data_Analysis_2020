```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
```

라인 플롯

```python
plt.plot([1,4,9,16]) # y좌표 ...x = [0,1,2,3] 이 디폴트
plt.title("Line Plot")
plt.show()
```

```python
plt.plot([1,4,9,16]) 
plt.title("Line Plot")
plt.grid(True)
plt.show()
```

```python
plt.plot([1,4,9,16]) # y좌표이다. ...x = [0,1,2,3] 이 디폴트
plt.title("Line Plot")
plt.grid(True)
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
plt.plot([10,20,30,40],[1,4,9,16]) 
plt.title("Line Plot")
plt.grid(True)
plt.show()
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
#한글사용
plt.plot([10,20,30,40],[1,4,9,16]) 
plt.title("X 틱스")
plt.xlabel('엑스축')
plt.ylabel('와이축')
plt.grid(True)
plt.show()
```

```python
font1 = {'family': 'NanumMyeongjo', 'size': 24, 
         'color':  'black'}
font2 = {'family': 'NanumBarunpen', 'size': 18, 'weight': 'bold',
         'color':  'darkred'}
font3 = {'family': 'NanumBarunGothic', 'size': 12, 'weight': 'light',
         'color':  'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.show()
```
