```python
x =np.arange(1.0,4.0,0.01)
y1 = x **2
y2 = x**2 /2

plt.plot(x,y1)
plt.plot(x,y2)
plt.title("두 그래프 사이 영역채우기")
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[100:201], y1[100:201],y2[100:201], color='lightgrey', alpha=0.5) 
plt.show()
```

```python
# 임의의 영역 채우기
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill([1.9, 1.9,2.5, 3.1, 3.1], [2, 5,9, 11, 8], color='lightgray', alpha=0.5)

plt.show()
```

그리드 축 지정하기

```python
#y 축만 지정, 축도 스타일 지정가능
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y',color ='red', alpha = 0.2, linestyle='--')

plt.show()
```

```python
#문제
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color='blue' ,label='sine')
plt.plot(x, y2, color='red', linewidth=2, label='cosine')


plt.xticks([0, np.pi/2,np.pi,3*np.pi/2,2*np.pi], labels=[r'$0$',r'$1/2\pi$',r'$1\pi$',r'$3/2\pi$',r'$2\pi$'])
plt.yticks(np.arange(-1, 1), labels=['low','zero','high'])
plt.title('sin_cosine')
plt.legend(loc = 'best') #loc = 0

plt.grid(True)

#수평선
```

파이 차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode,  shadow=True, colors=colors)
plt.show()
```

```python
#부채꼴
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
