```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
import nltk
from konlpy.tag import Okt
from konlpy.corpus import kobill
```

```python
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko[:1000]
```

```python
okt =Okt()
tokens_ko =okt.nouns(doc_ko)
tokens_ko[:10]
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
len(ko.tokens), len(set(ko.tokens))
```

```python
ko.vocab()
```

```python
plt.figure(figsize=(15,8))
ko.plot(50)
plt.show()
```

```python
stop_words= [ '.','(',')',',',"'",'%','-','X',').','x','의','자','에','안','번','호','을','이','다','만','로','가','를','및']
ko = [each_word for each_word in ko if each_word not in stop_words]
ko[:10]
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```

```python
#초등학교 는 몇번 언급되었는지
ko.count('초등학교')
```

```python
#원하는 단어의 문서내의 개략적위치와 분량 확인
plt.figure(figsize=(15,6))
ko.dispersion_plot(['육아휴직','초등학교','공무원'])
```

```python
#원하는 단어의 주변부 확인
ko.concordance('초등학교')
```

```python
#연어
ko.collocations()
```

```python
from wordcloud import WordCloud
data = ko.vocab().most_common(150)

wordcloud = WordCloud(#font_path='c/Wondows/Fonts/malgun.tff',
                    relative_scaling= 0.2,
                    background_color='white',
                    width=1000,height=600).generate_from_frequencies(dict(data))

plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
