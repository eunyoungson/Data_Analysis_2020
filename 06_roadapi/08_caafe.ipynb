```python
#스타벅스 17 'map-marker'
star = df_ganseogu.loc[df_ganseogu['상호명'].str.contains('스타벅스', na=False)]
star
```

```python
#이디야 39 'cloud'
edi = df_ganseogu.loc[df_ganseogu['상호명'].str.contains('이디야', na=False)]
#edi
```

```python
#빽다방 8 'gift'
bb = df_ganseogu.loc[df_ganseogu['상호명'].str.contains('빽다방', na=False)]
#bb
```

```python
#투썸 13 'info-sign' ts
ts = df_ganseogu.loc[df_ganseogu['상호명'].str.contains('투썸', na=False)]
#ts
```

```python
#엔제리너스 1 'pk-circle' angel
angel = df_ganseogu.loc[df_ganseogu['상호명'].str.contains('엔제리너스', na=False)]
#angel
```

```python
# 실험> 아이콘 같은상호끼리 지정해주기 - 스타벅스만 해보자.
```

```python
star['icon'] = 'map-marker'
star['color'] = 'green'
#star
```

```python
#까치산역점 확인
mapping = folium.Map(
    location=[37.532285,126.846036], zoom_start=12)

folium.Marker(
    location=[37.532285, 126.846036],
    popup='스타벅스',
    tooltip = '화곡점',
    icon=folium.Icon(color='green')
).add_to(mapping)

#mapping
```

```python
#스타벅스 전체 나타내보기 왜 14개지? 한개는 주소가 같은데..
mapping = folium.Map(
    location=[star['위도'].mean(),star['경도'].mean()], zoom_start=14)
for i in star.index:
    folium.Marker(
    location=[star['위도'][i], star['경도'][i]],
    popup = star['상호명'][i],
    #if star['지점명'][i]] == 'nan':
        #tooltip =star['상호명'][i]
    #else:
        #tooltip =star['지점명'][i],
    tooltip =star['지점명'][i],
    icon=folium.Icon(color=star.icon[i], icon=star.icon[i])
).add_to(mapping)
mapping
```

```python
#이디야
mapping = folium.Map(
    location=[edi['위도'].mean(),edi['경도'].mean()], zoom_start=14)
for i in edi.index:
    folium.Marker(
    location=[edi['위도'][i], edi['경도'][i]],
    popup = edi['상호명'][i],
    #if edi['지점명'][i]] == 'nan':
        #tooltip =edi['상호명'][i]
    #else:
        #tooltip =edi['지점명'][i],
    tooltip =edi['지점명'][i],
    icon=folium.Icon(color='red', icon='cloud')
).add_to(mapping)
#mapping
```

```python
# 5군데 카페 한지도에 표시하기 for문으로
mapping = folium.Map(
    location=[star['위도'].mean(),star['경도'].mean()], zoom_start=14)
for i in star.index:
    folium.Marker(
    location=[star['위도'][i], star['경도'][i]],
    popup = star['상호명'][i],
    #if star['지점명'][i]] == 'nan':
        #tooltip =star['상호명'][i]
    #else:
        #tooltip =star['지점명'][i],
    tooltip =star['지점명'][i],
    icon=folium.Icon(color='green', icon='map-marker')
).add_to(mapping)

for i in edi.index:
    folium.Marker(
    location=[edi['위도'][i], edi['경도'][i]],
    popup = edi['상호명'][i],
    #if edi['지점명'][i]] == 'nan':
        #tooltip =edi['상호명'][i]
    #else:
        #tooltip =edi['지점명'][i],
    tooltip =edi['지점명'][i],
    icon=folium.Icon(color='red', icon='cloud')
).add_to(mapping)

for i in bb.index:
    folium.Marker(
    location=[bb['위도'][i], bb['경도'][i]],
    popup = bb['상호명'][i],
    #if bb['지점명'][i]] == 'nan':
        #tooltip =bb['상호명'][i]
    #else:
        #tooltip =bb['지점명'][i],
    tooltip =bb['지점명'][i],
    icon=folium.Icon(color='blue', icon='cloud')
).add_to(mapping)

for i in ts.index:
    folium.Marker(
    location=[ts['위도'][i], ts['경도'][i]],
    popup = ts['상호명'][i],
    #if ts['지점명'][i]] == 'nan':
        #tooltip =ts['상호명'][i]
    #else:
        #tooltip =ts['지점명'][i],
    tooltip =ts['지점명'][i],
    icon=folium.Icon(color='orange', icon='cloud')
).add_to(mapping)

for i in angel.index:
    folium.Marker(
    location=[angel['위도'][i], angel['경도'][i]],
    popup = angel['상호명'][i],
    #if angel['지점명'][i]] == 'nan':
        #tooltip =angel['상호명'][i]
    #else:
        #tooltip =angel['지점명'][i],
    tooltip =angel['지점명'][i],
    icon=folium.Icon(color='black', icon='cloud')
).add_to(mapping)

mapping
```

```python
#아이콘과 색깔 컬럼 추가시켜서 데이터프레임의 컬럼수를 모두 통일 시킨다.
```

```python
star['icon'] = 'map-marker'; star['color'] = 'green'
edi['icon'] = 'cloud'; edi['color'] = 'red'
bb['icon'] = 'gift'; bb['color'] = 'blue'
ts['icon'] = 'info-sign'; ts['color'] = 'orange'
angel['icon'] = 'pk-circle'; angel['color'] = 'black'

```

```python
# 5군데 카페 한 데이터프레임으로 합치기
cafe = pd.concat([star,edi,bb,ts,angel], ignore_index=True)
print(cafe)
```
